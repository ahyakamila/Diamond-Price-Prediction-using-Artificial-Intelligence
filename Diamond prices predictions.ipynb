```python
df = df.drop(df[df["x"]==0].index)
df = df.drop(df[df["y"]==0].index)
df = df.drop(df[df["z"]==0].index)
df.shape
```

Mmebuang Outliers

```python
df = df[(df["depth"]<75)&(df["depth"]>45)]
df = df[(df["table"]<80)&(df["table"]>40)]
df = df[(df["x"]<30)]
df = df[(df["y"]<30)]
df = df[(df["z"]<30)&(df["z"]>2)]
df.shape
```

Representation Transformation

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
label_data = df.copy()
label_encoder = LabelEncoder()
for col in df[['cut','color','clarity']]:
    label_data[col] = label_encoder.fit_transform(label_data[col])
label_data.head()
```

```python
df=label_data
```

```python
df
```

Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_v = {}
max_v = {}
for i in ['carat','cut','color','clarity','depth','table','price','x','y','z']:
    min_v[i] = min(df[i])
    max_v[i] = max(df[i])
    df[i] = (df[i]-min(df[i]))/(max(df[i])-min(df[i]))
```

Data Partitioning

```python
from sklearn.model_selection import train_test_split
```

```python
X= df[['carat','cut','color','clarity','depth','table']]
y= df["price"]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=7)
```

MLP Regressor

```python
from sklearn.neural_network import MLPRegressor
```

```python
X_train
```

```python
y_train
```

```python
for i in range(10,21,10):
    for j in range(1,11):
        for k in ['adam','lbfgs','sgd']:
            regr = MLPRegressor(hidden_layer_sizes=(i),solver=k, learning_rate_init= j/100, max_iter=20, activation = "logistic", alpha = 0.0001, momentum = 0.9)
            regr.fit(X_train, y_train)
            print('Hidden layer ',i)
            print('learning rate ',j/100)
            print('solver ',k)
            regr.score(X_test,y_test)
            print(regr.score(X_test,y_test))
            print()
```

```python
regr = MLPRegressor(hidden_layer_sizes=(10),solver='adam', learning_rate_init= 0.08, max_iter=20, activation = "logistic", alpha = 0.0001, momentum = 0.9)
MLP = regr.fit(X_test, y_test)
```

K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
for i in range(1,6,1):
    for j in ['uniform', 'distance']:
        for k in ['auto', 'ball_tree', 'kd_tree', 'brute']:
            for l in (20, 41, 10):
                knn = KNeighborsRegressor(n_neighbors=(i), weights= j, algorithm = k, leaf_size = l)
                knn.fit(X_train, y_train)
                print('n_neighbors ',i)
                print('weights ',j)
                print(' algorithm  ',k)
                print('leaf_size ',l)
                knn.score(X_test,y_test)
                print(knn.score(X_test,y_test))
                print()
```

```python
knn = KNeighborsRegressor(n_neighbors=(5), weights= 'distance', algorithm = 'kd_tree', leaf_size = 10)
```
